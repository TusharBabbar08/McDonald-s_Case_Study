# Q4. Which category contributes to the maximum % of Cholesterol in a diet (% daily value)?

```python
df['Category'].nunique()
```

```python
sort=pd.DataFrame({'Sorted Category': sorted(df['Category'].unique())})
sort
```

```python
check = pd.pivot_table(df, 'Cholesterol (% Daily Value)', index=['Category'])


Result = check.sort_values(('Cholesterol (% Daily Value)'), ascending=False)

Result
```

### Above data shows a relation of the 9 Categories to the % Cholesterol.
### The data is sorted from 'High' to 'Low' as we move down in the categories

# Q5. Which item contributes maximum to the Sodium intake?

```python
check = pd.pivot_table(df, 'Sodium', index=['Item'])


Result = check.sort_values(('Sodium'), ascending=False)

Result.head(10)
```
